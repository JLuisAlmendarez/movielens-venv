```python
import pandas as pd
file_r = "/Users/josealmendarez/Downloads/ml-latest-small/movies.csv"

df = pd.read_csv(file_r)

df
```

```python
# Exploraci√≥n de datos (ED)
print("üìä Dimensiones del dataset:", df.shape)
print("\nüîé Primeras filas:")
print(df.head())

print("\nüìù Informaci√≥n general:")
print(df.info())

print("\nüìà Valores nulos por columna:")
print(df.isnull().sum())

print("\nüî¢ N√∫mero de valores √∫nicos por columna:")
print(df.nunique())

print("\nüé¨ Columnas disponibles:", df.columns.tolist())

# Estad√≠sticas descriptivas (solo si hubiera num√©ricas)
print("\nüìä Estad√≠sticas descriptivas:")
print(df.describe(include='all'))

# Distribuci√≥n de g√©neros
if "genres" in df.columns:
    genres = df["genres"].str.split("|").explode()
    print("\nüé≠ Top 10 g√©neros m√°s comunes:")
    print(genres.value_counts().head(10))
```

```python
# Distribuciones de variables num√©ricas
import matplotlib.pyplot as plt
import seaborn as sns

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in numeric_cols:
    plt.figure(figsize=(6,3))
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f"Distribuci√≥n de {col}")
    plt.show()

# Correlaciones entre num√©ricas
if len(numeric_cols) > 1:
    plt.figure(figsize=(8,6))
    sns.heatmap(df[numeric_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
    plt.title("üîó Matriz de correlaci√≥n")
    plt.show()

# An√°lisis de g√©neros (si existe la columna)
if "genres" in df.columns:
    genres = df["genres"].str.split("|").explode()
    top_genres = genres.value_counts().head(10)
    print("\nüé≠ Top 10 g√©neros m√°s comunes:\n", top_genres)
    
    plt.figure(figsize=(8,4))
    sns.barplot(x=top_genres.values, y=top_genres.index, palette="viridis")
    plt.title("üé≠ Top 10 g√©neros m√°s comunes")
    plt.xlabel("Frecuencia")
    plt.ylabel("G√©nero")
    plt.show()

# An√°lisis temporal (si hay timestamp o year)
if "year" in df.columns:
    plt.figure(figsize=(10,4))
    sns.countplot(x="year", data=df, order=sorted(df["year"].dropna().unique()))
    plt.title("üìÖ Distribuci√≥n de pel√≠culas por a√±o")
    plt.xticks(rotation=45)
    plt.show()

if "timestamp" in df.columns:
    df["date"] = pd.to_datetime(df["timestamp"], unit="s")
    plt.figure(figsize=(10,4))
    df["date"].dt.year.value_counts().sort_index().plot(kind="bar")
    plt.title("üìÖ N√∫mero de registros por a√±o")
    plt.show()
```
